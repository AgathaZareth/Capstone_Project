```python
import pandas as pd

# load dfs
MLB = pd.read_pickle("./MLB_df.pkl")
MiLB = pd.read_pickle("./MiLB_df.pkl")
College = pd.read_pickle("./College_df.pkl")
```

```python
# add `Level` column to all dfs
College['Level'] = 'Collegiate - Division_I'
MiLB['Level'] = 'Minor - Triple_A'
MLB['Level'] = 'Major'
```

```python
# concatenate dfs
baseball_hitting = pd.concat([MLB, MiLB, College], ignore_index=True)
```

```python
# check new df
baseball_hitting
```

```python
# check column types
baseball_hitting.info()
```

```python
# get column names, need to change col type for numerical cols
baseball_hitting.columns
```

```python
# grab just what should be numerical columns
num_cols = ['Games Played', 'At Bats', 'Runs', 'Hits', 'Doubles', 'Triples',
       'Home Runs', 'Runs Batted In', 'Walks', 'Strikeouts', 'Stolen Bases',
       'Caught Stealing', 'Batting Average', 'On-Base Percentage',
       'Slugging Percentage', 'On-Base Plus Slugging']
```

```python
# make a copy of df to work with 
df = baseball_hitting.copy()
```

```python
# change column types from strings (object) to floats
for col in num_cols:
    df = df.astype({col: float})
```

```python
# check column types 
df.info()
```

```python
# transpose .describe() for easy reading
df.describe().T
```

```python
# remove rbi's below 0 and compare with leaving them in 

# put in df to compare with leaving all rbi's in
temp_1 = pd.DataFrame(df[df['Batting Average']>0].describe()['Batting Average'])

# all .describe() with rbi's
temp_2 = pd.DataFrame(df.describe()['Batting Average'])

# compare .describe()'s
compare = pd.concat([temp_1, temp_2], axis=1)
compare.columns = ["below 0 rbi's REMOVED", "all rbi's"]
compare
```

```python
# drop players with rbi below 50 percentile (after rbi's at 0 are removed, see above `below 0 rbi's REMOVED`)
above_50_rbi = df[df['Batting Average']> 0.257]

# check the number of rows
print(above_50_rbi.shape)

# check number of remaining players at each level
above_50_rbi['Level'].value_counts()
```

```python
# list of all positions
positions = list(pd.DataFrame(df['Position'].value_counts()).index)
positions
```

```python
# who are top X number of players for each position?

# proof of concept with 1 player
def best_OPS_by_position(position, dataframe):#, df=baseball_hitting):
    if position in list(pd.DataFrame(dataframe['Position'].value_counts()).index):
    
        search_this_df = dataframe[dataframe['Position'] == position]
        best = search_this_df.sort_values(by='On-Base Plus Slugging', ascending=False).iloc[0]
        as_df = pd.DataFrame(best).T

    return as_df   
```

```python
DHS = df[df['Position'] == '1B']
best = DHS.sort_values(by='On-Base Plus Slugging', ascending=False).iloc[0]
pd.DataFrame(best).T
```

```python
DHS = above_50_rbi[above_50_rbi['Position'] == '1B']
best = DHS.sort_values(by='On-Base Plus Slugging', ascending=False).iloc[0]
pd.DataFrame(best).T
```

```python
best_OPS_by_position('SS', above_50_rbi)
```

```python
top_OPS_players = []
for position in positions:
    top_OPS_players.append(best_OPS_by_position(position, above_50_rbi))
    
pd.concat(top_OPS_players, ignore_index=True)
```
